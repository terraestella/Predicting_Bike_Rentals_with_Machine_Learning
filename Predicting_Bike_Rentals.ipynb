```python
import pandas as pd
import numpy as np

bike_rentals = pd.read_csv("bike_rental_hour.csv")
bike_rentals.head(24)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [9.5, 6]

bike_rentals['cnt'].hist()
plt.show()
```

```python
def assign_label(hour):
    if (hour >= 0) & (hour < 6):
        return 4
    elif (hour >= 6) & (hour < 12):
        return 1
    elif (hour >= 12) & (hour < 18):
        return 2
    elif (hour >= 18) & (hour <= 24):
        return 3
    
bike_rentals["time_label"] = bike_rentals["hr"].apply(assign_label)
```

```python
bike_rentals.head(24)
```

```python
train = bike_rentals.sample(frac=0.8)
test = bike_rentals.loc[~bike_rentals.index.isin(train.index)]
```

```python
print(train.shape)
print(test.shape)
```

Error metric will be MSE (mean squared error), because the data is continuous metric.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

predictors = list(train.columns)
predictors.remove("cnt")
predictors.remove("casual")
predictors.remove("registered")
predictors.remove("dteday")

lr = LinearRegression()
lr.fit(train[predictors], train["cnt"])
predictions = lr.predict(test[predictors])

lr_mse = mean_squared_error(test["cnt"], predictions)

lr_mse
```

```python
# Error is very high (probably due to some very high outliers in "cnt" column).
```

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor()
dtr.fit(train[predictors], train["cnt"])
dtr_predictions = dtr.predict(test[predictors])

dtr_mse = mean_squared_error(dtr_predictions, test["cnt"])

dtr_mse
```

```python
dtr = DecisionTreeRegressor(min_samples_leaf=5)
dtr.fit(train[predictors], train["cnt"])
dtr_predictions = dtr.predict(test[predictors])

dtr_mse = mean_squared_error(dtr_predictions, test["cnt"])

dtr_mse
```

```python
dtr = DecisionTreeRegressor(min_samples_leaf=7)
dtr.fit(train[predictors], train["cnt"])
dtr_predictions = dtr.predict(test[predictors])

dtr_mse = mean_squared_error(dtr_predictions, test["cnt"])

dtr_mse
```

```python
from sklearn.ensemble import RandomForestRegressor

rfg = RandomForestRegressor()
rfg.fit(train[predictors], train["cnt"])
rfg_predictions = rfg.predict(test[predictors])

rfg_mse = mean_squared_error(rfg_predictions, test["cnt"])

rfg_mse
```

```python
from sklearn.ensemble import RandomForestRegressor

rfg = RandomForestRegressor(min_samples_leaf=2)
rfg.fit(train[predictors], train["cnt"])
rfg_predictions = rfg.predict(test[predictors])

rfg_mse = mean_squared_error(rfg_predictions, test["cnt"])

rfg_mse
```

```python
from sklearn.ensemble import RandomForestRegressor

rfg = RandomForestRegressor(n_estimators=150, min_samples_leaf=2)
rfg.fit(train[predictors], train["cnt"])
rfg_predictions = rfg.predict(test[predictors])

rfg_mse = mean_squared_error(rfg_predictions, test["cnt"])

rfg_mse
```
